```python
encoded_df = pd.get_dummies(df, columns=['Gender','Vehicle_Age','Vehicle_Damage'], drop_first=False).astype(int)
encoded_df_test= pd.get_dummies(df_test, columns=['Gender','Vehicle_Age','Vehicle_Damage'], drop_first=False).astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

# Assume encoded_df and encoded_df_test are already defined

X = encoded_df[['Age', 'Driving_License', 'Region_Code', 'Previously_Insured', 'Annual_Premium', 'Policy_Sales_Channel', 
        'Vintage', 'Gender_Female', 'Gender_Male', 'Vehicle_Age_1-2 Year', 'Vehicle_Age_< 1 Year', 'Vehicle_Age_> 2 Years', 
        'Vehicle_Damage_No', 'Vehicle_Damage_Yes']]
y = encoded_df['Response']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the KNN model
knn = KNeighborsClassifier(n_neighbors=5)  # You can choose a different number of neighbors

# Fit the model on the training data
knn.fit(X_train, y_train)

# Predict on the test data
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Generate classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)

# Ensure the test dataset has the same feature columns as the training dataset
X_test_final = encoded_df_test[['Age', 'Driving_License', 'Region_Code', 'Previously_Insured', 
                                'Annual_Premium', 'Policy_Sales_Channel', 'Vintage', 'Gender_Female', 
                                'Gender_Male', 'Vehicle_Age_1-2 Year', 'Vehicle_Age_< 1 Year', 
                                'Vehicle_Age_> 2 Years', 'Vehicle_Damage_No', 'Vehicle_Damage_Yes']]

# Make predictions on the test data
y_test_pred = knn.predict(X_test_final)

# Prepare the output DataFrame
output_df = pd.DataFrame({
    'id': encoded_df_test['id'],
    'Response': y_test_pred
})

# Save the output DataFrame to a CSV file
output_df.to_csv('predictions.csv', index=False)
```
